Q6) Calculate the Expected number of candies for a randomly selected child Below are the probabilities of count of candies for children (ignoring the nature of the child-Generalized view)

Child A – probability of having 1 candy = 0.015 Child B – probability of having 4 candies = 0.20 Child C – probability of having 3 candies = 0.65 Child D – probability of having 5 candies = 0.005 Child E – probability of having 6 candies = 0.01 Child F – probability of having 2 candies = 0.120

```python
#ANSWER: Expected number of candies for a randomly selected child is :

Probability=[0.015,0.20,0.65,0.005,0.01,0.120]

No_of_Candies=[1,4,3,5,6,2]

x=0

for i in Probability:
    for j in No_of_Candies:
        if Probability.index(i)==No_of_Candies.index(j):
            x=x+(i*j)
            
print(x)
```

Q 20) Calculate probability from the given dataset for the below cases Calculate the probability of MPG of Cars for the below cases. MPG <- Cars$MPG a. P(MPG>38) b. P(MPG<40) c. P (20<MPG<50)

```python
car=pd.read_csv("C:\\Users\\This\\Documents\\ExcelR\\Cars (1).csv")
car_mpg = car['MPG']
car_mpg_mean = car.MPG.mean()
car_mpg_std = car.MPG.std()
print(car_mpg_mean)
print(car_mpg_std)

#P(MPG>38)
print('Probability for MPG>38 is:', 1-stats.norm.cdf(38,car_mpg_mean,car_mpg_std))

#P(MPG<40)
print('Probability for MPG<40 is:', stats.norm.cdf(40,car_mpg_mean,car_mpg_std))

#P(20<MPG<50)
mpg_20 = stats.norm.cdf(20,car.MPG.mean(),car.MPG.std())
mpg_50 = 1-stats.norm.cdf(50,car.MPG.mean(),car.MPG.std())
#print(mpg_20)
#print(mpg_50)
print('Probability for 20<MPG<50 is:',mpg_50-mpg_20)
```

Q 24) A Government company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days Hint: rcode  pt(tscore,df) df  degrees of freedom

```python
#H0= average life >= 260 days
#H1 = average life < 260 days
# tscore = x - µ / (ϭ / √n)

tscore = (260-270) / (90/ np.sqrt(18))
print('T_Score is', tscore)
p_value = stats.t.cdf(-0.4714, df = 17)
print('P_value is:', p_value)
```
